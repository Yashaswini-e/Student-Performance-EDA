```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df['G3'], bins=10)
plt.xlabel('Final Grade (G3)')
plt.ylabel('Number of Students')
plt.title('Distribution of Final Grades')
plt.show()
```

â€¢ The distribution of final grades shows that most students score in the mid-range,
  indicating average overall academic performance across the dataset.


```python
plt.figure(figsize=(6,4))
sns.boxplot(x='studytime', y='G3', data=df)
plt.xlabel('Study Time Level')
plt.ylabel('Final Grade (G3)')
plt.title('Study Time vs Final Grade')
plt.show()
```

â€¢ Students with higher study time levels tend to have higher median final grades,
  suggesting a positive relationship between study time and academic performance.


```python
df[['G1', 'G2', 'G3']].mean().plot(kind='bar')
plt.xlabel('Grade Type')
plt.ylabel('Average Score')
plt.title('Average Grades Progression')
plt.show()
```

â€¢ The average grades across G1, G2, and G3 show a consistent progression, indicating
  that student performance remains relatively stable over time.
